#  Dashboard Redes Sociais

##  1. Dashboard Geral da Conta (Visão Executiva)

**Objetivo:** Visão macro de como a conta está performando.  
**Principais métricas:**
- Seguidores atuais e crescimento semanal/mensal
- Alcance total vs. Impressões 
- Taxa de engajamento média 
- Cliques no site e visitas ao perfil

---

##  2. Dashboard de Conteúdo (Posts e Reels)

**Objetivo:** Entender quais tipos de posts geram mais engajamento.  
**Principais métricas:**
- Lista de publicações com:
  - Tipo (imagem, carrossel, vídeo, reels)
  - Curtidas, comentários, salvos, compartilhamentos
  - Alcance e impressões
  - Taxa de engajamento por post
- Top 5 posts mais engajados (ranking)
- Comparação entre tipos de mídia (gráfico de barras empilhadas)

---

##  3. Dashboard de Stories

**Objetivo:** Medir como os stories mantêm a audiência atenta.  
**Principais métricas:**
- Alcance vs. impressões por story (gráfico de linhas ou dispersão)
- Métricas de interação (respostas, toques para avançar/voltar, saídas)
- Identificar horários em que stories performam melhor

---

##  4. Dashboard Demográfico (Público-Alvo)

**Objetivo:** Conhecer quem segue a PyLadies Floripa.  
**Principais métricas:**
- Distribuição por faixa etária e gênero (gráficos de pizza ou barras)
- Distribuição por país e cidade (mapa ou tabela)
- Crescimento de seguidores ao longo do tempo

---

##  5. Dashboard de Atividade

**Objetivo:** Descobrir os melhores horários e dias para postar.  
**Principais métricas:**
- Dias e horários de maior atividade dos seguidores
- Comparação de engajamento por dia da semana

#  Dashboard Geral da Conta - PyLadies Floripa

##  Objetivo
Fornecer uma **visão macro da performance da conta**, consolidando métricas principais em um único painel para acompanhamento rápido.

---

##  Métricas-Chave
| Métrica | Descrição | Fonte (API) |
|---------|-----------|-------------|
|  **Seguidores Atuais** | Número total de seguidores no momento | `/account` |
|  **Crescimento de Seguidores** | Variação semanal ou mensal de seguidores | `/insights` (follower_count ao longo do tempo) |
|  **Alcance Total (Reach)** | Número de contas únicas alcançadas no período | `/insights` (métrica `reach`) |
|  **Impressões (Impressions)** | Número total de visualizações | `/insights` (métrica `impressions`) |
|  **Taxa de Engajamento** | (Curtidas + Comentários + Salvos) ÷ Alcance | Calculada manualmente a partir de `/media` |
|  **Cliques no Site** | Total de cliques no link do perfil | `/insights` (métrica `website_clicks`) |
|  **Visitas ao Perfil** | Quantidade de visitas ao perfil | `/insights` (métrica `profile_views`) |

---

 **Filtros úteis:**
- Período (últimos 7, 14 ou 30 dias)
- Tipo de mídia (opcional, se quiser cruzar com dados de conteúdo)

---

##  Dados Necessários
- Endpoint `/me/accounts` → para pegar ID da conta.
- Endpoint `/account` → para dados básicos (seguidores, etc).
- Endpoint `/insights` → métricas de alcance, impressões, profile_views, website_clicks.
- Endpoint `/media` → para somar curtidas, comentários, salvos e calcular engajamento.

# Dashboard de Conteúdo – PyLadies Floripa

## Objetivo
Permitir analisar o desempenho de cada tipo de conteúdo publicado (posts e reels), identificando quais formatos e temas geram mais engajamento.

---

## Métricas-Chave por Publicação
| Métrica | Descrição | Fonte (API) |
|---------|-----------|-------------|
| **Tipo de Mídia** | Imagem, Vídeo, Carrossel, Reel | `/media` |
| **Legenda** | Texto associado à publicação | `/media` |
| **Data de Publicação** | Data/hora em que foi postada | `/media` |
| **Curtidas** | Total de curtidas | `/media/{media-id}/insights` |
| **Comentários** | Total de comentários | `/media/{media-id}/insights` |
| **Salvos** | Quantidade de vezes que salvaram o post | `/media/{media-id}/insights` |
| **Compartilhamentos** | Quantidade de compartilhamentos (parcial) | `/media/{media-id}/insights` |
| **Alcance** | Contas únicas alcançadas | `/media/{media-id}/insights` |
| **Impressões** | Visualizações totais | `/media/{media-id}/insights` |
| **Taxa de Engajamento** | (Curtidas + Comentários + Salvos) ÷ Alcance | cálculo manual |

---

## Dados Necessários
- Endpoint `/media` para listar todas as publicações com ID, tipo e legenda.
- Endpoint `/media/{media-id}/insights` para métricas detalhadas (likes, comentários, alcance, impressões, salvos, etc).
- Calcular a **taxa de engajamento** manualmente com base nas métricas obtidas.
